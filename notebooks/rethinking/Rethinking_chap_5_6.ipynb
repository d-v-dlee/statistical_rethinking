```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Here we are sampling from the prior using the following model definition:

$y_{i}$ ~ Normal( $\mu$, $\sigma$)  #likelihood <br>
$\mu$ ~ Normal(0, 10)  # prior for mean<br>
$\sigma$ ~ Uniform(0, 10)  # prior for stdev <br>

```python
sample_mu = np.random.normal(0, 10, size=1000)
sample_sigma = np.random.uniform(0, 10, size=1000)
prior_h = np.random.normal(sample_mu, sample_sigma, size=1000)
sns.set_style('whitegrid')
sns.kdeplot(prior_h, bw=0.5);
```
